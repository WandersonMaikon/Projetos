```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/Wande/DATA SCIENSE/focos48h_AMS.csv')
```

```python
dataframe = data.tail()
print(dataframe)
```

```python
estados = data[data['pais'] == 'Brasil']['estado'].unique()
print(estados)
```

```python
municipios = data[data['estado'] == 'RONDÔNIA']['municipio']
print(municipios)
```

```python
horario = data[data['estado'] == 'RONDÔNIA'][['data_hora_gmt','municipio']]
print(horario)
```

```python
contagem_estados = data[data['pais'] == 'Brasil']['estado'].value_counts()
print(contagem_estados)
```

```python
plt.bar(contagem_estados.index, contagem_estados.values)
plt.xlabel('Estado')
plt.ylabel('Contagem')
plt.title('Contagem de Estados')
plt.xticks(rotation=90)
plt.show()
```
