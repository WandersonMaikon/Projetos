```python
ano = [1950, 1960,1970, 1980, 1990, 2000, 2010]
gpd = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
plt.plot(ano, gpd, color='green',marker='o',linestyle='solid')
#adicinando Titulo Gráfico eixo x Ano, y gpd
plt.title('PIB Nominal')
#Adicionando legenda
plt.ylabel('Bilhões de dolares')
plt.show()
```
